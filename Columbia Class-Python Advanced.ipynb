15.Pivot Table

```python
#Single Index
table = pd.pivot_table(data=df,index=['term'])
table
```

```python
#multiple indexes
table = pd.pivot_table(df,index=['term','grade'])
table
```

```python
#multiple indexes with different aggregate functions

table = pd.pivot_table(df,index=['term','grade'],aggfunc={'annual_inc':np.mean,'CLI':np.sum})
table
```

```python
table = pd.pivot_table(df,index=['int_rate'],columns=['grade'],values=['CLI'],aggfunc=np.mean)
table
```
